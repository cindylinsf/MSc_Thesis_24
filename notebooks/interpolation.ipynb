```python
# Server Management
def start_comfyui_server():
    """Start ComfyUI server and open browser interface."""
    try:
        process = subprocess.Popen(
            ['python', str(COMFY_BASE / 'main.py')],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True  # This will give us text output instead of bytes
        )
        
        # Print the first few lines of output to see what's happening
        print("Server output:")
        while True:
            output = process.stdout.readline()
            if not output:
                break
            print(output.strip())
            if "To see the GUI go to:" in output:
                webbrowser.open('http://127.0.0.1:8188')
                break
        
        return process
        
    except Exception as e:
        print(f"Failed to start ComfyUI server: {str(e)}")
        raise
```

```python
import os
import sys
import subprocess
from pathlib import Path

# Define base path
COMFY_BASE = Path("/Users/cindylinsf/Documents/CCI/THESIS/Msc_Thesis_Project_Files/models/ComfyUI")

# Define global server variable
global_server = None

def start_comfyui_server():
    global global_server
    try:
        process = subprocess.Popen(
            [sys.executable, str(COMFY_BASE / 'main.py')],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        global_server = process
        print("Server started")
        return process
    except Exception as e:
        print(f"Failed to start ComfyUI server: {e}")
        raise

def restart_comfyui():
    global global_server
    
    # Stop existing server if it's running
    if global_server is not None:
        global_server.terminate()
        print("Stopped existing server")
    
    # Clean any potential cache
    cache_dir = COMFY_BASE / "__pycache__"
    if cache_dir.exists():
        import shutil
        shutil.rmtree(cache_dir)
        print("Cleared cache")
    
    # Start new server
    global_server = start_comfyui_server()
    print("\nServer restarted. Please:\n1. Wait 30 seconds\n2. Open http://127.0.0.1:8188 in a new browser window")

# Restart the server
restart_comfyui()
```
